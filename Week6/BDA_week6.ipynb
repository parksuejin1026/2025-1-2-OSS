```python
df = pd.DataFrame({'A' : range(1, 6),
                   'B' : range(6,-3, -2),
                   'C' : range(10, 5, -1),
                   'D clmns' : range(3, -2, -1),
                   'E' : ['python', 'java', 'python', 'go', 'python']})
df
```

```python
df.query('A > B')
```

```python
df[df.A > df.B]
```

```python
df.query('E == "python"')
```

```python
df.query("E != 'python'")  # E 열의 인덱스가 python이 아니면 전체행 출력
```

```python
df.query('C == 10')
```

```python
df.query('A > 0')
```

```python
df.query(' A > 3 | B > 8' )
```

```python
df.query(' A > 2 & C < 8 ')
```

```python
df.query(" E in ['python', 'go']")
```

```python
df.query('A == "D clmns"')  # 문자열에 빈칸이 있으면 df로 출력
```
